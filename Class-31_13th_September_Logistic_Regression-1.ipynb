# Logistic Regression

Logistic Regression is one of the supervised machine learning algorithms.  
It is mainly used for **classification** problems—specifically **binary classification**.

### Q: What is Binary?

**Answer:**  
Binary means there are only two possible outcomes, such as:
- 0 or 1  
- Pass or Fail  
- True or False  

For example, predicting whether a person is **COVID positive or negative**, or whether a student will **pass or fail**.  

To predict such categories (two possible classes), we use **Logistic Regression**.

# How it works?

Why is it again called **regression**?  
If it is a **classification**, why are they calling it regression?  

It is because it also uses the same equation but in a different way.

\$[
y = mx + c
\$]

# Why Not Use Linear Regression for Classification?

One common interview question is:  
**“If we want to predict binary data (yes/no, true/false, positive/negative), why can’t we just use Linear Regression?”**

### Answer:  
Imagine you go to a hospital and ask a doctor, based on your symptoms (like cough, cold, fever), whether you are **COVID positive or not**.  

- The doctor (acting as a model) must give you a **clear answer**: **positive** or **negative**.  
- He cannot give you something "in-between," like **0.67 COVID**.  

Linear Regression, however, predicts **continuous values** (e.g., 0.2, 0.67, 1.5, etc.), which are **not suitable for classification**.  

That’s why we use **Logistic Regression**—because it converts the continuous values into **probabilities between 0 and 1**, and then classifies them into categories.  


# Logistic Regression Uses `mx + c` – How It Works

Logistic Regression still uses the **linear equation**:

\[
y = mx + c
\]

But instead of using the raw output for prediction, it applies a **Sigmoid Function** to convert the continuous output into probabilities.

## Behind the Scenes

The **Sigmoid Function**, also called the **logit function**, is defined as:

\$[
\sigma(z) = \frac{1}{1 + e^{-z}}
\$]



- Suppose the linear part (`mx + c`) gives a value, e.g., `0.59`.  
- The Sigmoid function maps **any real number** into a value between **0 and 1**.  
- If the result is **≥ 0.5**, we classify it as **positive (1)**.  
- If the result is **< 0.5**, we classify it as **negative (0)**.

## Sigmoid Function – Role

- Converts **continuous numbers** (from the linear equation) into **probabilities**.  
- Helps classify outputs into **binary categories** like:  
  - 0 or 1  
  - Pass or Fail  
  - Positive or Negative  

This is the **core idea behind Logistic Regression**: take a linear output and convert it into **classes** using the sigmoid function.

# Evaluation of a Classification Model

Now, let's understand **how to evaluate a classification model**.  

### Linear Regression Metrics

In **Linear Regression**, common metrics include:  
- MAE (Mean Absolute Error)  
- MSE (Mean Squared Error)  
- RMSE (Root Mean Squared Error)  
- R² Score  

These metrics measure **how good or bad a regression model is**, but **they are not suitable for classification problems**.  

> **Important:** For classification, we **cannot mix regression metrics with classification metrics**.  

### Classification Evaluation Metrics

For classification models, we use **different metrics** to evaluate performance:  

- **Accuracy** – How many predictions were correct overall  
- **Recall** – Ability to find all positive cases  
- **Precision** – How many predicted positives are actually positive  
- **F1 Score** – Harmonic mean of precision and recall  
- **Specificity** – Ability to correctly identify negatives  
- **AUC (Area Under Curve)** – Measures model’s ability to distinguish classes  
- **ROC (Receiver Operating Characteristic curve)** – Plots True Positive Rate vs False Positive Rate  

> To fully understand these metrics, we first need to understand the **Confusion Matrix**.

# Confusion Matrix

Let's assume **100 people visited a hospital**, and each person was given a test based on:  

- Glucose Level  
- Blood Pressure (BP)  
- Sugar Levels  

These tests give **numerical values**, and the doctor predicted whether each person is **positive or negative** for a condition **before the actual reports arrived**.  

After two days, the actual reports are in, and now we can compare the **doctor's predictions** with the **actual results**.  

### Q: How Good Was the Doctor's Prediction?

To evaluate, we compare the **predicted results** with the **actual outcomes**:  

- **Correct Prediction** → 100% accuracy for that case  
- **Wrong Prediction** → 0% accuracy for that case  

We can summarize this comparison in a **table**, which is exactly what the **Confusion Matrix** does.  

![conf.png](conf.png)

### Confusion Matrix Terms

- **TP (True Positive)** – Predicted positive **and** actually positive  
- **TN (True Negative)** – Predicted negative **and** actually negative  
- **FP (False Positive)** – Predicted positive **but** actually negative  
- **FN (False Negative)** – Predicted negative **but** actually positive  

The **credibility of the model** depends on **how many correct predictions it made**.

> The Confusion Matrix is the foundation for computing all other classification metrics, such as Accuracy, Precision, Recall, F1 Score, etc.


# Accuracy

**Mathematical Formula:**

\$[
\text{Accuracy} = \frac{TP + TN}{TP + TN + FP + FN}
\$]

**Explanation:**  
Accuracy is defined as the **total number of correct predictions** divided by the **total number of predictions**.  

It measures how often the model **classifies correctly** overall.

### Example:

Suppose we tested **100 patients** for a disease:  

| Outcome        | Predicted Positive | Predicted Negative |
|----------------|-----------------|-----------------|
| Actual Positive | 40 (TP)         | 10 (FN)         |
| Actual Negative | 5 (FP)          | 45 (TN)         |

- **TP = 40**  
- **TN = 45**  
- **FP = 5**  
- **FN = 10**  

\$[
\text{Accuracy} = \frac{TP + TN}{TP + TN + FP + FN} = \frac{40 + 45}{40 + 45 + 5 + 10} = \frac{85}{100} = 0.85
\$]

 **Accuracy = 85%**  

This means the model correctly predicted **85 out of 100 patients**.

# Recall (Sensitivity)

**Mathematical Formula:**

\$[
\text{Recall} = \frac{TP}{TP + FN}
\$]

**Explanation:**  
Recall (also called **Sensitivity**) measures the model’s ability to correctly identify **positive cases**.  
It tells us, out of all actual positives, **how many were correctly predicted** by the model.  

Recall focuses **only on positive outcomes**, showing the model's effectiveness in identifying them.

### Example:

Suppose a model predicts cancer in patients:  

- **True Positives (TP) = 50** → Correctly identified cancer patients  
- **False Negatives (FN) = 200** → Missed cancer patients  

\$[
\text{Recall} = \frac{TP}{TP + FN} = \frac{50}{50 + 200} = \frac{50}{250} = 0.2
\$]

 **Recall = 20%**  

This means the model correctly identified **only 20% of the actual cancer patients**.

# Precision

**Mathematical Formula:**

\$[
\text{Precision} = \frac{TP}{TP + FP}
\$]

**Explanation:**  
Precision measures, **out of all predicted positive cases**, how many were **actually positive**.  
It tells us the model’s **accuracy in predicting positives**.

### Example:

Suppose a model predicts cancer in patients:  

- **True Positives (TP) = 50** → Correctly predicted cancer patients  
- **False Positives (FP) = 100** → Incorrectly predicted as cancer  

\$[
\text{Precision} = \frac{TP}{TP + FP} = \frac{50}{50 + 100} = \frac{50}{150} \approx 0.33
\$]

 **Precision = 33%**  

This means that **only 33% of the patients predicted as cancer actually have cancer**.

# F1 Score

Sometimes we face a problem:  

> "We have Accuracy, Precision, and Recall. Which one should we consider? Which is better?"  

The answer depends on the **type of problem** we are solving.  
Different metrics may give **different insights**, so we need a **balanced measure** when there is a trade-off.

### Trade-off Concept

- Precision focuses on **correct positive predictions**.  
- Recall focuses on **capturing all actual positives**.  
- F1 Score **balances both Precision and Recall** to give a single metric.  

Think of it as:  
> Neither completely Precision nor completely Recall — **staying in the middle**.  

### F1 Score Definition

F1 Score is the **harmonic mean of Precision and Recall**.  

\$[
\text{F1 Score} = 2 \times \frac{\text{Precision} \times \text{Recall}}{\text{Precision} + \text{Recall}}
\$]

It helps us evaluate a model when **both Precision and Recall matter**.

### Example:

From previous examples:  

- **Precision = 0.33**  
- **Recall = 0.2**  

\$[
\text{F1 Score} = 2 \times \frac{0.33 \times 0.2}{0.33 + 0.2} = 2 \times \frac{0.066}{0.53} \approx 0.25
\$]

 **F1 Score = 0.25**  

This means the model achieves a **balanced performance** considering both Precision and Recall.

# Why Recall and Precision Matter

Understanding **Recall** and **Precision** is important because they have **real-world impact**. The choice of metric depends on the **problem you are trying to solve**.

![err.png](err.png)

## Recall Example (Healthcare)

- **Scenario:** 100 people visit a hospital.  
- **Prediction:** The doctor/model predicts 15 people as **negative** for a disease (e.g., COVID).  
- **Reality:** Those 15 people are actually **positive**.  

**Impact:**  
- These 15 people may unknowingly spread the disease while thinking they are safe.  
- This is an example of a **Type II Error (False Negative)**.  

**Insight:**  
- A **high recall** is critical in healthcare because **missing actual positives** can have serious consequences.

## Precision Example (Investment)

- **Scenario:** Investment experts advise a businessman to invest in a particular area.  
- **Prediction:** The experts predict it will be profitable (**positive**).  
- **Reality:** No customers come, and the investment loses money (**False Positive**).  

**Impact:**  
- Low precision means that **many predicted positives are actually wrong**, which can lead to financial loss.  

**Insight:**  
- Precision matters when **false positives are costly**.  

> **Key Point:**  
The **choice of metric** depends on the context:  
- Use **Recall** when missing positives is dangerous (healthcare, safety).  
- Use **Precision** when false positives are costly (finance, legal decisions, spam detection).

# Specificity (True Negative Rate)

**Definition:**  
Specificity measures **how well the model identifies actual negatives**.  
It answers the question: _“Out of all actual negative cases, how many did the model correctly predict as negative?”_

**Mathematical Formula:**

\$[
\text{Specificity} = \frac{TN}{TN + FP}
\$]

**Explanation:**  
- TN = True Negatives (correctly predicted negatives)  
- FP = False Positives (incorrectly predicted positives)  

Specificity quantifies the **proportion of correctly identified negatives** out of all actual negatives.

### Example:

Suppose we tested **100 patients** for a disease:  

| Outcome        | Predicted Positive | Predicted Negative |
|----------------|-----------------|-----------------|
| Actual Positive | 40 (TP)         | 10 (FN)         |
| Actual Negative | 5 (FP)          | 45 (TN)         |

- **TN = 45**  
- **FP = 5**  

\$[
\text{Specificity} = \frac{TN}{TN + FP} = \frac{45}{45 + 5} = \frac{45}{50} = 0.9
\$]

 **Specificity = 90%**  

**Interpretation:** The model correctly identified **90% of the actual negative patients**.

### False Positive Rate (FPR)

\$[
\text{FPR} = 1 - \text{Specificity} = \frac{FP}{TN + FP} = \frac{5}{50} = 0.1
\$]

**Interpretation:** 10% of actual negatives were incorrectly predicted as positive.


# AUC (Area Under the Curve)

Sometimes we have a problem:  

> "We tried different classification models, and each has its own ROC curve. Which model should we pick?"  

### Solution:

We use **AUC (Area Under the Curve)** to compare them.

- **AUC** tells us how good a model is at **separating positives and negatives**.  
- **Higher AUC** = better model.  
- AUC values range from **0 to 1**:  
  - **1** → Perfect model  
  - **0.5** → Model is like random guessing  
  - **0** → Worst model  

### Example:

Suppose we have **two models** to predict if a patient has a disease:  

| Model | AUC  | Comment                  |
|-------|------|--------------------------|
| A     | 0.85 | Good at distinguishing positive and negative cases |
| B     | 0.65 | Not as good               |

 **We should choose Model A** because it has the **higher AUC**, meaning it predicts positives and negatives more accurately across all thresholds.

```python
# Let's distribute the data in every column

plt.figure(figsize=(20,15), facecolor='blue')
graph=1
for column in data:
    if graph <= 9:
        plt.subplot(3,3, graph)
        sns.distplot(data[column])
        plt.xlabel(column)
    graph += 1
plt.show()
```

We can see that the data has some skewness, so let’s handle it.

Also, some columns like **Glucose, Insulin, Skin Thickness, BMI, and Blood Pressure** contain values equal to 0. This is not possible in reality (for example, a person cannot have 0 Glucose or 0 Blood Pressure).

To fix this, we have two options:

- 1: Remove those rows with 0 values.

- 2: Replace the 0 values with the mean of that column.

We will go with the second option (replacing with the mean).

```python
# Replacing with the mean

data['BMI']=data['BMI'].replace(0,data['BMI'].mean())
data['Insulin']=data['Insulin'].replace(0,data['Insulin'].mean())
data['BloodPressure']=data['BloodPressure'].replace(0,data['BloodPressure'].mean())
data['Glucose']=data['Glucose'].replace(0,data['Glucose'].mean())
data['SkinThickness']=data['SkinThickness'].replace(0,data['SkinThickness'].mean())

# Try again:
plt.figure(figsize=(20,15), facecolor='green')
graph=1
for column in data:
    if graph <= 9:
        plt.subplot(3,3, graph)
        sns.distplot(data[column])
        plt.xlabel(column)
    graph += 1
plt.show()
```
