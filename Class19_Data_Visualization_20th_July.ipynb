## Difference between Matplotlib and Seabord

Both are popular python libraries used for data visualization

### Matplotlib

1. Basic Library: Matplotlib is the foundation of all python libraries. It gives full control overy every part of a figure.
2. Low-level: You have to write more code to customize plots. Everything must be built step-by-step.
3. Versatile: You can create many types of plots - line, bar, scatter, histogram etc... but sometimes styling is manual.
4. More Customization: You can fine tune every part of the chart - fonts, color, ticks, levels, grids etc...

### Seaborn

1. Built on Matplotlib: Seaborn is based on Matplotlib but gives prettier and simpler charts with less code.
2. High-level API: It is easier to make complex plots like boxplots, violin plots and heatmaps.
3. Beautiful by default: Comes with better styling, color themes, and design out of the box.
4. Works well with DataFrames: Seaborn directly works with pandas dataframes, making it great for statistical plot.

#### Conclusion

- Use **Matplotlib** when you need full control and customization
- User **Seaborn** when you want quick, beautiful and statistical visualization

### Boxplot

It displays Outliers. 

```python
# Let's try to create a boxplot with the above dataframe
plt.figure(figsize=(4,4))
sns.boxplot(data=df)
plt.yticks(range(5,60,5))  #Range on Y axis
plt.xlabel('Age and Salary')
plt.show()
```

Notice 5 lines in both the plots. Let's concentrate on Orange plot starting from below there are total 5 horizontal lines including the surface of orange box.

- First line: **minimum value**
- Second line (below surface of orange box): 25%
- Third line (line within the box): 50%
- Fourth line (upper surface of orance box): 75%
- Fifth line: **maximum value**

**NOTE**: Any data lies between minimum and maximum value, it is a normal data. Anything outside this boundary are known as **outliers**.

### How it finds the Outliers?

Let's understand how it works in the backend to find the outliers. It uses **Inter Quantile Range (IQR)** for calculating.

```python
# Let's take a list

sal=[15,22,24,25,30,23,34,43,19,20,52,21,2]
sal.sort()
sal
```

```python
# Convert sal into series
import pandas as pd
ser=pd.Series(sal)
ser
```

```python
# While calculating IQR, we need below 3 values

# 1st quantile
q1=ser.quantile(0.25)
print("Q1",q1)

# 3rd quantile
q3=ser.quantile(0.75)
print("Q3",q3)

# Find the IQR for outliers
iqr=q3-q1
print("IQR",iqr)
```

Given the dataset:
[15, 19, 20, 21, 22, 23, 24, 25, 30, 34, 43, 52]

n=Total elements in the dataset = 12

### Percentile Calculations

â€¢ **25th percentile:**

(25/100) (n + 1) = 0.25 (12 + 1) = 3.25
The data at the 3.25th position is approximately 20.5 (interpolated between the 3rd and 4th elements). 3.25 is the index value.

â€¢ **50th Percentile (Median):**

(50/100) * (n + 1) = 0.5 * (12 + 1) = 6.5
The data at the 6.5th position is approximately 23.5 (interpolated between the 6th and 7th elements).

â€¢ **75th Percentile:**

(75/100) * (n + 1) = 0.75 * (12 + 1) = 9.75
The data at the 9.75th position is approximately 31.5 (interpolated between the 9th and 10th elements).

### Outlier detection formula

higher side --> Q3 + (1.5*IQR)

lower side --> Q1 - (1.5*IQR)

```python
upperside=q3+(1.5*iqr)
print("1. Anything which is more than:",upperside,"is an outlier.")
lowerside=q1-(1.5*iqr)
print("2. Anything which is lesser than:",lowerside,"is an outlier.")
```
