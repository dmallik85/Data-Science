### Jointplot Analysis: Total bill vs. tip

This is jointplot which shows the relationship between two numeric variables along with their individual distributions.

#### What does the main scatter plot show?

 X-axis: total bill (Total bill amount)
 Y-axis: tip (Tip amount)
 Each blue dot represents one customer.
The plot shows how tip amount varies with the total bill.

#### Trend Observed:

As the total bill increases, the tip generally increases too.
 However, it's not a perfect straight-line relationship - there's a positive correlation, but it's not very strong.

#### What about the histograms?

 Top histogram: Distribution of total bill
    - Most bills are around 10-20.
    
 Right histogram: Distribution of tip
    - Most tips are around 2-4.

# Basics of Plotting

Now, let's create a basic plot and learn how to customize it with labels, axes and headings.

```python
# Generate some random numbers
x=np.linspace(0,10,100)
y=np.sin(x)
```

```python
# Create a new figure
plt.figure(figsize=(8,6))

# Plot the data
plt.plot(x,y)

# Add lebels and title
plt.xlabel("X-axis of the plot")
plt.ylabel("Y-axis of the label")
plt.title("sin waveform")

# Show the plot
plt.show()
```

## Customizing the plot

```python
plt.figure(figsize=(12,7))

# Plot the data with a thicker, dashed red line
plt.plot(x,y,color='red',linewidth=2,linestyle='--',label='sin(x)')

# Add a Legend
plt.legend()

# Customize the x and y axis
plt.xlim(0,10)   # Set the range of x-axis
plt.ylim(-1.5,1.5)   # Set the range of y-axis

# Add grid lines
plt.grid(True,linestyle=':')

# Add labels and title with custom fonts
plt.xlabel("X-axis",fontsize=14)
plt.ylabel("Y-axis",fontsize=14)
plt.title("Customized Sine Wave Plot",fontsize=16,fontweight='bold')

# Add text annotation
plt.text(6.9,0.5,"sin(x) function",fontsize=16)

plt.show()
```

# Using Seaborn for statistical visualization

Seaborn is built on top of Matplotlib and provides a high-level interface for drawing attractive statistical graphics. Let's create a simple visualization using seaborn:

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
data=np.random.randn(100,2)    # It will generate 100 rows and 2 columns
# Create a scatter plot using seaborn
plt.figure(figsize=(10,6))
sns.scatterplot(x=data[:,0],y=data[:,1])   # x: 0th row to last and column 0; y: 0th row to last and column 1.
plt.xlabel('X-axis',fontsize=12)
plt.ylabel('Y-axis',fontsize=12)
plt.title('Seaborn Scatter Plot',fontsize=4)
plt.show()
```

## Kernel Density Estimation (KDE)

It is a way to estimate the probability distribution of a dataset like a smooth version of a histogram.

In simple words, KDE draws a smooth curve to show where your data points are most concentrated.

**Why Use KDE?**

- A histogram uses bars to show data distribution.
- KDE gives a smooth curve insted of bars.
- It helps to see patterns and trends more clearly.

Example:

```python
import pandas as pd
data={
    'Group A':np.random.normal(0,1,1000),    # mean, standard deviation, number of values
    'Group B':np.random.normal(2,1.5,1000),
    'Group C':np.random.exponential(2,1000)  # Generates 1000 random numbers from an exponential distribution with scale parameter 2
}
df=pd.DataFrame(data)
df
```

```python
import seaborn as sns
plt.figure(figsize=(15,5))
for i, column in enumerate(df.columns,1):  # Enumerate iterates with variable i as column
    plt.subplot(1,3,i)   # Multiple graph in single plot, 1 row, 3 plots and i = index wise
    sns.histplot(df[column],kde=True)
    plt.title(f'Histogram of {column}',fontsize=14)
    plt.xlabel('Value',fontsize=10)
    plt.ylabel('Fequency',fontsize=10)
plt.tight_layout()
plt.show()
```

Histograms help us visualize:

- The overall shape of the distribution (e.g, normal, skewed)
- The central tendency
- The spread of the data
