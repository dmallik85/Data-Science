# Supervised Machine Learning

In **Supervised Machine Learning (SML)**, the model learns from past data (training data) to predict future data.  
Both **input** (features) and **output** (labels/targets) are provided during training, which makes it "supervised".

### Key Points:
- **Input (X):** Features / Independent variables  
- **Output (Y):** Labels / Class / Target – Dependent variable  

### Types of Supervised Learning:
1. **Regression**  
   - Used when predicting **continuous values** (e.g., salary, temperature, price).  
   - Example: Predicting house prices.  

2. **Classification**  
   - Used when predicting **categories or classes** (e.g., Yes/No, Male/Female, Spam/Not Spam).  
   - Example: Predicting whether an email is spam or not.  

### Example:
If your client gives you a dataset (like CSV or tabular data) containing both **input features** and **output labels**, then it is a **Supervised Machine Learning problem**.

# Regression  

In statistics, regression is the process of predicting a **label** (dependent variable) using the **features** (independent variables) we have.  
It is often used for time series modeling, finding cause-and-effect relationships between variables, and making forecasts.  

**Example:**  
We can study the relationship between a company's stock price and factors like customer reputation or annual performance using regression.

Regression analysis is an important tool for analyzing and modeling data.  
Here, we fit a curve or a line to the data points so that the difference between the actual data points and the curve/line is as small as possible.

# Linear Regression  

Linear Regression is one of the most basic and well-known machine learning algorithms, and it is often the first one people learn.  

**Main parts of a linear regression model:**  
- Discrete or continuous **independent variable(s)** (features)  
- A **best-fit regression line**  
- Continuous **dependent variable** (label)  

A linear regression model predicts the dependent variable using a regression line based on the independent variables.  

## Formula:  

\$[
Y = A + B \times X + E
\$]  

or  

\$[
Y = M \times X + C
\$]  

If you know the values of **M**, **X**, and **C**, then **Y** is your prediction.  

Where:  
- **A** = Intercept (value of Y when X = 0)  
- **B** or **M** = Slope of the line (how much Y changes when X changes)  
- **E** = Error term (difference between predicted and actual values)  
- **C** = Intercept (in the second form of the equation)  

---

## Slope (m) and Intercept (c) formulas  

For a simple linear regression line: (Least Square Method)

\$[
m = \frac{ \sum (x_i - \bar{x})(y_i - \bar{y}) }{ \sum (x_i - \bar{x})^2 }
\$]  

\$[
c = \bar{y} - m \times \bar{x}
\$]  

Where:  
- \$(m\$) = slope of the line  
- \$(c\$) = intercept of the line  
- \$(x_i, y_i\$) = data points  
- \$(\bar{x}, \bar{y}\$) = mean of x and y respectively  

---

**How it works:**  
1. **Features** → Independent Variables  
2. **Labels** → Dependent Variables  

**Equation used:**  

\$[
y = a + b \times x + e
\$]

Here e refers to 'Error'

# Simple Linear Regression  

In simple linear regression, we use **only one feature** to predict the label (target variable).  

It is a method for predicting a **quantitative response** (continuous variable) using a single input variable (feature).  

## Mathematical Equation:  

\$[
y = c + m \times x
\$]  

or  

\$[
y = \beta_0 + \beta_1 x
\$]  

Where:  
- \$( y \$) = Predicted value (label)  
- \$( c \$) or \$( \beta_0 \$) = Intercept (value of \$( y \$) when \$( x = 0 \$))  
- \$( m \$) or \$( \beta_1 \$) = Slope (how much \$( y \$) changes for a one-unit change in \$( x \$))  
- \$( x \$) = Input variable (feature)

# What do terms represent?

Y is the response or the target variable
X is the feature
𝛽1 is the coefficient od x
𝛽2 is the intercept.

𝛽0 and 𝛽1 are the model coefficients. To create a model, we must "learn" the values of these coefficients. And once we have the value of these
coefficients, we can use the model to predict the Sales!

Estimating ("Learning") Model Coefficients
The coefficients are estimated using the least-squares criterion, i.e., the best fit line has to be calculated that minimizes the sum of squared
residuals (or "sum of squared errors").

```python
#create x=feature and y=label (simple linear regression: 1 feature and 1 label)

x=data[['TV']]   # Saving independent variable 'TV' into x. This is one way to derive in simple liner regression
y=data.sales     # Saving dependent variable 'sales' into y. This is another way to derive in simple liner regression

#follow the usual sklearn patter : import , instantiate ,fit


from sklearn.linear_model import LinearRegression
lm=LinearRegression()
lm.fit(x, y)
```

```python
# Print intercept and co-efficient

print(lm.intercept_)    # This is c in y=mx+c formula
print(lm.coef_)         # This is m in y=mx+c formula
```
