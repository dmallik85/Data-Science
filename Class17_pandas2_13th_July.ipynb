## Operating with NULL values

```python
# Create dataframe with missing values
import pandas as pd
import numpy as np
raw_data={'first_name':['Jason',np.nan,'Tina','Jake','Amy','Madhu'],
          'last_name':['Millar',np.nan,'Ali','Milner','Cooze','Patil'],
          'age':[42,np.nan,36,24,73,25],
          'sex':['M',np.nan,'F','M','F','F'],
          'unit-1':[4,np.nan,np.nan,2,3,4],
          'unit-2':[3,np.nan,np.nan,4,4,4]}
#df=pd.DataFrame(raw_data)
df=pd.DataFrame(raw_data,columns=['first_name','last_name','age','sex','unit-1','unit-2'])    #If you want to specify particular column in a dataframe
df
```

```python
df_no_missing=df.dropna()    #It will remove rows with NaN values as asxis is not specified. By default it considers as axis=0 if not specified.
df_no_missing
```

```python
df
```

```python
# How to remove rows that have all the NULL values
df_cleaned=df.dropna(how='all')    #This formula will drop the rows if there are all null values in a row
df_cleaned
```

```python
# Create a column with full of missing values
df['finally_score']=np.nan
df
```

```python
# Drop Null values column wise where all are Null in a colum
df_nan_col=df.dropna(axis=1,how='all')
df_nan_col
```

## Fillna

```python
# Filling NaN with 0 wherever present
df.fillna(0)
```

```python
# Let's see how to fill NaN values with a mean of other values from a particular column
df['unit-1'].fillna(df['unit-1'].mean(),inplace=True)
df
```

```python
# Fill in the missing in unit-2 column with each sex's mean value of unit-2
df['unit-2'].fillna(df.groupby('sex')['unit-2'].transform('mean'),inplace=True)
df
```

```python
# How to backfill -it means, Tina's value will be filled in place of NaN values from row 1, just as an example
df.fillna(method='bfill')
```

```python
# How to frontfill -it means, Jason's value will be filled in place of NaN values from row 1, just as an example.
df.fillna(method='ffill')
```

```python
# What if we want to fille the column 'finally_score' with addition of unit-1 and unit-2
df['finally_score'].fillna(df['unit-1']+df['unit-2'],inplace=True)
df
```

```python
# Adding a column 'Location' with value 'Bangalore'
df['Location']='Bangalore'
df
```

## Hierarchical Indexing (Multiindexing)

Till now we have gone through:
- 1-dimensional data (like series)
- 2-dimensional data (like DataFrame)

But some time we have to work with the data where more than 2 levels of indexing - for example, data organized by two or more keys (like 'State' and 'City' together). In such cases, it's better to user hierarchical indexing or multiindexing, where:

- We can use multiple index level in a single index
- This helps us handle higher dimensional data using regular series or dataframe objects.

### What to learn in this section?

1. How to create multiindex object directly
2. How to index, slice and compute using multiindex
3. How to convert between normal and multiindex formats

```python
import pandas as pd

raw_data={'company': ['Google', 'Google', 'Google', 'Google', 'Facebook', 'Facebook', 'Facebook', 'Facebook', 'Amazon', 'Amazon', 'Amazon', 'Amazon'],
          'project': ['1st', '1st', '2nd', '2nd', '1st', '1st', '2nd', '2nd', '1st', '1st', '2nd', '2nd'],
          'incharge': ['sunder', 'ruth', 'benjamin', 'serger', 'mark', 'shery', 'eduardo', 'david', 'jeffrey', 'andy', 'keith', 'edilth'],
          'sales(in million)': [54,24,31,12,23,24,34,31,32,23,52,13],
          'profit(in million)':[14,8,9,4,10,11,14,13,12,7,22,4]}
df=pd.DataFrame(raw_data, columns=['company', 'project', 'incharge', 'sales(in million)', 'profit(in million)'])
df
```

```python
# In the above dataframe, there is already an index value there. However, if we would like to make 'Google' or 'Project' as index, how to perform that?
df.shape
```

```python
df.set_index(['company','project','incharge'],drop=False)     # drop=False means, it doesn't drop those columns while making them index
```

```python
df.set_index(['company','incharge'],drop=True,inplace=True)
df
```

```python
# Summarize the result of a company
df['company'].sum()
df
```
